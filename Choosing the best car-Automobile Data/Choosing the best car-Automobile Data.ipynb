```python
#create new columns
#now create a seperate column for average milage

auto['dummy']=1
auto['avg_mpg']=(auto['city_mpg']+auto['highway_mpg'])/2
auto.head()
```

```python
#remove columns
auto.drop(['avg_mpg','dummy'],axis=1,inplace=True)
auto.head()
```

```python
#add again those columns and use diff method to
#remove those columns
auto['dummy']=1
auto['avg_mpg']=(auto['city_mpg']+auto['highway_mpg'])/2

auto=auto.drop(['avg_mpg','dummy'],axis=1)
auto.head()
```

```python
#removing a row
#checking for the row names

print(auto.index)
auto1=auto.drop([0,1],axis=0)
auto1.head()
```

```python
auto.describe()
```

```python
#lets select a part of the datframe and work with them

auto2=auto.loc[0:10,:]
auto2
```

```python
#description of the column city milage
auto2.city_mpg.describe()
```

```python
#we see that min milage is 17 and max is 24,lets check for all the cars with milage>20
auto2.city_mpg>20
```

```python
#get all the details of those selected cars
auto2[auto2.city_mpg>20]
```

```python
#getting only the makers of the car
auto2[auto2.city_mpg>20].make
```

```python
#always remember to pass the column name with a list otherwise the name of the col wont display
auto2[auto2.city_mpg>20][['city_mpg']]
```

```python
#the makers and its milage
auto2[auto2.city_mpg>20][['make','city_mpg']]
```

```python
#getting all those columns which we require
auto2[auto2.city_mpg>20][['make','city_mpg','engine_location','horsepower']]
```

```python
#generally cars with more hp have low milage, lets see their correlation
auto2[['city_mpg','horsepower']].corr()
```

```python
#description of the horsepower column
auto2.horsepower.describe()
```

```python
#getting the mean horsepower and mean city milage

print(auto2.city_mpg.mean(),'\n',auto2.horsepower.mean())
```

```python
#checking for cars with hp>115 and milage>20
#we want better milage as well as good power

auto[auto.horsepower>115][auto.city_mpg>20]
```

```python
#the above list are our required cars
#lets check their bodystyles and also count them

print(auto[auto.horsepower>115][auto.city_mpg>20][['body_style']],'\n')
auto[auto.horsepower>115][auto.city_mpg>20]['body_style'].value_counts()
```
