 Merging Dataframes

```python
import pandas as pd

#creating a dataframe
df = pd.DataFrame([{'Name': 'Lilendar', 'Bought from flipkart': 'Notebooks', 'Cost': 120},
                   {'Name': 'Ajay', 'Bought from flipkart': 'Pen', 'Cost': 200},
                   {'Name': 'Vignesh', 'Bought from flipkart': 'Spoon', 'Cost': 90}],
                  index=['Store 1', 'Store 2', 'Store 3'])
df
```

```python
#adding the date column
df['Date'] = ['December 1', 'January 1', 'mid-May']
df
```

```python
#delivey status
df['Delivered'] = True
df
```

```python
#providing the feedback details
df['Feedback'] = ['Positive', None, 'Negative']
df
```

```python
#creating a diff dataframe and using the new-index as keys to change dates
#we have to change every rows if changing , otherwise NaN is filled
df1 = df.reset_index()
df1['Date'] = pd.Series({0: 'December 1', 2: 'mid-May'})
df1
```

```python
adf = df.reset_index()
adf['Name'] = pd.Series({0: 'piyush', 2: 'Martin'})
adf
```

```python
#creating 2 seperate dataframes
staff_df = pd.DataFrame([{'Name': 'piyush', 'Role': 'Director of HR'},
                         {'Name': 'dipti', 'Role': 'Course liasion'},
                         {'Name': 'sarah', 'Role': 'Grader'}])
staff_df = staff_df.set_index('Name')

student_df = pd.DataFrame([{'Name': 'piyush', 'School': 'Business'},
                           {'Name': 'dipti', 'School': 'Law'},
                           {'Name': 'reema', 'School': 'Engineering'}])
student_df = student_df.set_index('Name')
print(staff_df)
print()
print(student_df.head())
```

```python
#all the observations will be captured
pd.merge(staff_df, student_df, how='outer', left_index=True, right_index=True)
```

```python
#only the intersection will be captured
pd.merge(staff_df, student_df, how='inner', left_index=True, right_index=True)
```

```python
#here left and right can be thought of two circles intersecting each other
#ist dataframe is theleft circle and the 2nd dataframe is the right circle
#only the observation of the 1st dataframe 
pd.merge(staff_df, student_df, how='left', left_index=True, right_index=True)
```

```python
#observation of the 2nd data frame
pd.merge(staff_df, student_df, how='right', left_index=True, right_index=True)
```

```python
staff_df1 = staff_df.reset_index()
student_df1 = student_df.reset_index()
pd.merge(staff_df1, student_df1, how='left', left_on='Role',right_on='Name')
```

```python
#creating a new dataframe
df = pd.DataFrame(['A+', 'A', 'B+', 'B', 'B-', 'C+', 'C'],
                  index=['excellent', 'good', 'ok', 'below avg', 'poor', 'bad', 'very bad'])
df.rename(columns={0: 'Grades'}, inplace=True)
df
```

```python
#head of the categorical datatypes
df['Grades'].astype('category')
```

```python
grades = df['Grades'].astype('category',
                             categories=['D', 'D+', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+'],
                             ordered=True)
grades.head()
```

Pivot Tables

```python
import numpy as np
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.pivot_table(values='(kW)', index='YEAR', columns='Make')
```

```python
#creating a pivot table we need 3 things,values+index+columns
#here the mean value is imputed from all the values of a particular year
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=np.mean)
```

```python
#using both mean and min values for imputataion
df.pivot_table(values='(kW)', index='YEAR', columns='Make', 
               aggfunc=[np.mean,np.min],margins=True)
```

```python
df.pivot_table(values='(kW)', index='YEAR', columns='Make', aggfunc=[np.mean,np.min])
```
